**Iris Flower Classifaction using SVC and Logistic Regression models**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sea
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Load the dataset
df = pd.read_csv('Iris.csv')
```

```python
# Extract features (x) and target labels (y)
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
# Visualize the data
sea.pairplot(df,hue='Species')
plot.title("Plot of Iris Dataset")
plot.show()
```

```python
# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

**Dataset is divided such that 75% is used for model training while 25% is given for model testing.**

```python
# Create an instance of the SVC model
svc_model = SVC()
```

```python
# Fit the model to the training data
svc_model.fit(x_train, y_train)
print (x_train)
```

```python
# Predict the labels for the test data
y_predict = svc_model.predict(x_test)
```

```python
# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_predict)
print("Accuracy of SVC Model is:", accuracy)
```

```python
# Train the Logistic Regression model
log_model = LogisticRegression()
log_model.fit(x_train, y_train)
```

```python
# Predict using the Logistic Regression model
y_predict_log = log_model.predict(x_test)
log_accuracy = accuracy_score(y_test, y_predict_log)
print("Accuracy of Logistic Regression is:", log_accuracy)
```

**Thus we are getting 94.7% accuracy using SVC model while 100% accuracy using Logistic Regression model.**
